```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
% matplotlib inline
from kmodes import kmodes
```

```python
data_1 = pd.read_csv('Nomis Solutions_Data Part I.csv')
data_2 = pd.read_csv('Nomis Solutions Part II.csv')
```

```python
data_1.columns = ['tier', 'fico', 'approve_date', 'term', 'amount', 'previous_rate',
       'car_type', 'competition_rate', 'outcome', 'rate', 'cost_of_funds',
       'partner_bin']
```

```python
data_2 = data_2.drop(['Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14',
                      'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17'], axis=1)
data_2 = data_2.drop(0,axis=0)
data_2.columns = data_1.columns
data_2.reset_index(inplace=True)
data_2.drop('index',axis=1,inplace=True)
```

```python
data = data_1.merge(data_2,how='outer')
```

```python
amount_list = [float(i) for i in data.amount]
rate_list = [float(i)/100 for i in data.rate]
term_list = [float(i)/12 for i in data.term]
```

```python
interest_amounts = [round(x*y*z,2) for x,y,z in zip(amount_list,rate_list,term_list)]
```

```python
data['interest'] = interest_amounts
```

```python
sold = data[data['outcome'] == 1]
lost = data[data['outcome'] == 0]
percentage = len(sold)/len(lost)
percentage # 28.2% of applicants took the loan.
```

```python
sold_date_list = list(sold['approve_date'].unique())
sold.approve_date = pd.to_datetime(pd.Series(sold.approve_date))
sold.approve_date = sold.approve_date.dt.to_period('M')

interest_month = sold[['approve_date']+['interest']]
interest_month_sums = interest_month.groupby('approve_date').sum()
interest_month_sums.reset_index(inplace=True)

interest_month_sums.plot(x='approve_date', y='interest', figsize=(16,12), yticks=[0,2000000,4000000,6000000,8000000,10000000,12000000,14000000])
```

```python
approved = data.drop('approve_date', axis=1)
approved = approved[data['outcome'] == 1]

approved['tier']=approved['tier'].apply(float)
approved['fico']=approved['fico'].apply(float)
approved['term']=approved['term'].apply(float)
approved['amount']=approved['amount'].apply(float)
approved['competition_rate']=approved['competition_rate'].apply(float)
approved['outcome']=approved['outcome'].fillna(0).apply(int).apply(str)
approved['rate']=approved['rate'].apply(float)
approved['cost_of_funds']=approved['cost_of_funds'].apply(float)
approved['partner_bin']=approved['partner_bin'].apply(float).apply(str)
approved['interest']=approved['interest'].apply(float)

fico_amount_rate = approved[['fico']+['amount']+['rate']].values.tolist() #1
fico_rate_comp = approved[['fico']+['rate']+['competition_rate']].values.tolist() #2
fico_amount_int = approved[['fico']+['amount']+['interest']].values.tolist() #3
fico_int_cost = approved[['fico']+['interest']+['cost_of_funds']].values.tolist() #4 
#rate_comp_outcome = approved[['rate']+['competition_rate']+['outcome']].values.tolist() #5 
#fico_comp_rate = approved[['fico']+['competition_rate']+['rate']].values.tolist() #6
```

```python
km = kmodes.KModes(n_clusters=6, init='Huang', n_init=3, verbose=1)
```

```python
clus2_one = km.fit_predict(fico_amount_rate)
print(km.cluster_centroids_)
```

```python
clus2_two = km.fit_predict(fico_rate_comp)
print(km.cluster_centroids_)
```

```python
clus2_three = km.fit_predict(fico_amount_int)
print(km.cluster_centroids_)
```

```python
clus2_four = km.fit_predict(fico_int_cost)
print(km.cluster_centroids_)
```
